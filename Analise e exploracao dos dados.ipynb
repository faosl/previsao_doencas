```python
atendimentos_df = pd.read_csv('atendimentos.csv',sep=';',parse_dates=True)
```

```python
atendimentos_df.set_index('SEQ_REQUISICAO_ITEM',inplace=True)
```

```python
classe_exame_rotina = 'Z00'
nome_coluna_classes = 'GRUPO_CID'
```

```python
capitulos_cids = ['A00-B99',
                  'D50-D89',
                  'E00-E90',
                  'G00-G99',
                  'J00-J99',
                  'K00-K93',
                  'M00-M99',
                  'I00-I99']

atendimentos_df = atendimentos_df[ ((atendimentos_df['COD_ABRAGENCIA'].isin(capitulos_cids)) | 
                        ((atendimentos_df[nome_coluna_classes]== classe_exame_rotina)))]
```

```python
df = resultados_df.set_index('SEQ_REQUISICAO_ITEM',drop=False)\
                  .join(atendimentos_df,how='inner')
```

```python
classes_gp = df[nome_coluna_classes].value_counts()
classes_gp = classes_gp[classes_gp>=200]
tops_classes = list(classes_gp.index)
classes_gp
```

```python
df_classes = pd.get_dummies(df[nome_coluna_classes])
classes_colunas = list(df_classes.columns)
df = pd.concat([df,df_classes],axis=1)
```

```python
df.drop(['SEQ_REQUISICAO',
                 'FLG_SEXO',
                 'DAT_NASCIMENTO',
                 'FLG_GESTANTE',
                 'SEQ_CID',
                 'COD_ABRAGENCIA',
                 'DES_CAPITULO',                 
                 'COD_CID',
                 'NOM_CID',
                 'SEQ_REQUISICAO_ITEM'],axis=1,inplace=True)
```
